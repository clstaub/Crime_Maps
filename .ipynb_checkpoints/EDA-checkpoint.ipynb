```python
def date_split(string):
    dtobj = datetime.datetime.strptime(string, '%Y-%m-%d %H:%M:%S')
    return dtobj.date()
```

```python
df['date'] = df['OCCURRED_ON_DATE'].apply(lambda x: date_split(x))
```

```python
def time_split(string):
    dtobj = datetime.datetime.strptime(string, '%Y-%m-%d %H:%M:%S')
    return dtobj.time()
```

```python
df['time'] = df['OCCURRED_ON_DATE'].apply(lambda x: time_split(x))
```
